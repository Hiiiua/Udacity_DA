**Task 2**: Let's look at the relationship between two other numeric variables. How does the engine size relate to a car's CO2 footprint? The 'displ' variable has the former (in liters), while the 'co2' variable has the latter (in grams per mile). Use a heat map to depict the data. How strong is this trend?

```python
y_bin = np.arange(fuel_econ['co2'].min(), fuel_econ['co2'].max()+50, 50)
x_bin = np.arange(fuel_econ['displ'].min(), fuel_econ['displ'].max()+0.4, 0.4)
```

```python
h2d = plt.hist2d(data = fuel_econ, y='co2', x='displ', bins = [x_bin, y_bin], cmin=0.5, cmap='viridis_r')
plt.colorbar()
```

```python
h2d = plt.hist2d(data = fuel_econ, y='co2', x='displ', bins = [x_bin, y_bin], cmin=0.5, cmap='viridis_r')
plt.colorbar()
counts = h2d[0] #返回图像上的非na色块
# for i in range
# counts.shape[0]#要计数的dimension

#遍历 算数
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j] #取出色块对应得频数
        if c>=300:
#             plt.text(1,11, int(c), ha='center', va='center', color='white',clip_on=True)

            plt.text(x_bin[i]+0.2, y_bin[j]+25, int(c), ha='center', va='center', color='white',clip_on=True)
        elif c > 0:
            plt.text(x_bin[i]+0.2, y_bin[j]+25, int(c), ha='center', va='center', color='black',clip_on=True)
```
